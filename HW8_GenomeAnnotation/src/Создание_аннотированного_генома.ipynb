# Создание аннотированного генома

##Добавление annotations

```python
scaffolds = {} #пустой словарь

#аннотация
for scaffold in SeqIO.parse("/content/scaffolds.fa", "fasta"):
  scaffold.annotations['molecule_type'] = 'DNA'
  scaffold.annotations['organism'] = 'T. oleivorans Barents'
  
  #добавляем аннотированную запись в словарь
  scaffolds[scaffold.id] = scaffold 
```

## Добавляем координаты генов

```python
#рассмотрим структуру файла с координатами
!head -n 20 gms2.lst
```

```python
!tail -n -20 gms2.lst
```

```python
coordFile = open('gms2.lst', 'r')
id = ""
coords = []
features = []
for line in coordFile:
  if line[0] == '#' or not line.strip(): #пропускаем строчки с информацией и пустые
    continue

  elif line[:10] == 'SequenceID': #узнаём какой скаффолд
    if id != "":
      scaffolds[id].features = features
    id = line[12:].split("\n")
    id = id[0]
    features = []

  else: #координаты для скаффолда
    #достаём данные из строки
    coords = line.split()

    start = coords[2]
    if start[0] == '<':
      start = int(start[1:])
    else:
      start = int(start)

    end = coords[3]
    if end[0] == '>':
      end = int(end[1:])
    else:
      end = int(end)

    if coords[1] == '+':
      strnd = 1
    else:
      strnd = -1

    locus_tag = str(coords[0])

    #создаём feature
    feature = SeqFeature(FeatureLocation(start, end, strand=strnd), type="CDS")
    feature.qualifiers['locus_tag'] = [locus_tag]
    features.append(feature)

scaffolds[id].features = features #так в цикле сначала идёт нахождение нового скаффолда, а затем добавление features в предыдущий, то для последнего скаффолда, нужно вызвать ещё раз
```

## Добавляем трансляции генов (аминокислотные последовательноти)

```python
for id in scaffolds:
  for feature in scaffolds[id].features:
    for gene in SeqIO.parse("proteins.fasta", "fasta"):
      if gene.id == feature.qualifiers['locus_tag'][0]:
        feature.qualifiers['translation'] = [gene.seq]
```

## Добавляем функции белков (из бактерии MIL-1)

```python
#выведем данные
names = ['qseqid', 'sseqid', 'pident', 'length', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore']
mil1hits = pd.read_csv('scaffolds.hits_from_MIL_1.txt', sep='\t', header=None, names=names)
mil1hits
```

```python
#выберем с наименьшим evalue и уберём повторяющиеся
hits = mil1hits[mil1hits['sseqid'].str.contains("CCU")].sort_values('evalue', ascending=True).drop_duplicates('qseqid')
hits = hits.sort_values('qseqid', ascending=True)
hits
```

```python
for id in scaffolds:
  for feature in scaffolds[id].features:
    for line in hits.iterrows():
      if int(line[1]['qseqid']) == int(feature.qualifiers['locus_tag'][0]):
        feature.qualifiers['note'] = [line[1]['sseqid'].split('_')[2]]
```

```python
mil_1_genome = SeqIO.read("/content/T_oleivorans_MIL_1.gbk", "genbank")
```

```python
for  mil_f in mil_1_genome.features:
  if 'protein_id' not in mil_f.qualifiers:
    continue
  if 'product' not in mil_f.qualifiers:
    continue

  for id in scaffolds:
    for feature in scaffolds[id].features:
      if ('note' in feature.qualifiers):
        if feature.qualifiers['note'][0] == mil_f.qualifiers['protein_id'][0]:
          feature.qualifiers['product'] = mil_f.qualifiers['product']
```

## Добавляем функции белков (из БД SwissProt)

Читаем информацию из файла scaffolds.hits_from_SwissProt.txt

```python
swisshits_raw = pd.read_csv('scaffolds.hits_from_SwissProt.txt', sep='\t', header=None, names=names)
swisshits = swisshits_raw.sort_values('evalue', ascending=True).drop_duplicates('qseqid')
swisshits = swisshits.sort_values('qseqid',ascending=True)
swisshits
```

```python
for id in scaffolds:
  for feature in scaffolds[id].features:
    for line in swisshits.iterrows():
      if ('note' not in feature.qualifiers):
        if int(line[1]['qseqid']) == int(feature.qualifiers['locus_tag'][0]):
          feature.qualifiers['note'] = [line[1]['sseqid'].split('|')[2]]
      else:
        if int(line[1]['qseqid']) == int(feature.qualifiers['locus_tag'][0]):
          feature.qualifiers['note'].append(line[1]['sseqid'].split('|')[2])
```

Получить функции белков SwissProt из файла: https://ftp.uniprot.org/pub/databases/uniprot/current_release/knowledgebase/complete/uniprot_sprot.dat.gz

```python
!grep '^ID\|^DE   RecName: Full=' uniprot_sprot.dat > SwissProt_names.txt
```

```python
previd = None
swissgenes = dict()
for line in open('SwissProt_names.txt'):
  if line.startswith('ID'):
    previd = line.split()[1]
  if line.startswith('DE'):
    swissgenes[previd] = line.split('=')[1][:-2]
```

Добавляем к нашим features

```python
for id in scaffolds:
  for feature in scaffolds[id].features:
    for noteSP in swissgenes:
      if ('note' in feature.qualifiers):
        if feature.qualifiers['note'][0] == noteSP:
          feature.qualifiers['product'] = [swissgenes[noteSP]]
        elif len(feature.qualifiers['note']) == 2:
          if feature.qualifiers['note'][1] == noteSP:
            feature.qualifiers['product'].append(swissgenes[noteSP])
```
