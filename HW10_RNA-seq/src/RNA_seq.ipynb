# Выравнивание чтений

* Для каждого образца запускаем FastQC. Также через multiqc получаем объединенный .html файл.



```python
names = ["SRR3414629", "SRR3414630", "SRR3414631", "SRR3414635", "SRR3414636", "SRR3414637"]
for name in names:
  !./FastQC/fastqc /content/rna_seq_2021/{name}_1.fastq
```

```python
!multiqc .
```

* Запускаем HISAT2 и картируем все чтения на геном мыши.
* Общее кол-во чтений, и кол-во, которое удалось успешно откартировать (уникально и не уникально), указано в файле названиеОбразца.hisat.

```python
for name in names:
  !time hisat2 -p 3 -x mm10/genome -U /content/rna_seq_2021/{name}_1.fastq -S {name}.sam  2>  {name}.hisat
  !rm -v /content/rna_seq_2021/{name}_1.fastq
```

* Отбираем только те чтения, которые откартировались уникально (флаг NH:i:1):

```python
for name in names:
  !grep -P '^@|NH:i:1$' {name}.sam > {name}.uniq.sam
```

* Удаляем исходные .sam файлы, т.к. они нам больше не потребуются:

```python
for name in names:
  !rm -v {name}.sam
```

* Считаем кол-во уникально-картированных чтений


```python
for name in names:
  print(name)
  !grep -v '^@' {name}.uniq.sam | wc -l
  print()
```

* С помощью программы `HTSeq` Подсчитываем количество чтений, попавших на каждый ген:
* Аннтоация генома мыши в формате .gtf находится в файле `gencode.vM25.annotation.gtf`


```python
for name in names:
    !time htseq-count --format=sam --stranded=no {name}.uniq.sam gencode.vM25.annotation.gtf > {name}.counts
    !rm -v {name}.uniq.sam
```

* Смотрим сколько чтений не удалось приписать ни одному гену:
* `__no_feature` – столько чтений соответствует участкам генома, где не аннотировано ни одного экзона
* `__ambiguous` – столько чтений могут принадлежать разным генам
* Более подробно про такие случаи можно посмотреть в описании HTSeq-count - https://htseq.readthedocs.io/en/release_0.11.1/count.html
* Итого, общее число чтений, соответствующих хотя бы одному гену равно:   <br>**количествоУникальноКартированныхЧтений – __no_feature – __ambiguous = ...**


```python
for name in names:
    print(name)
    !grep '^__no_feature' {name}.counts
    !grep '^__ambiguous' {name}.counts
    print()
```

* Объединям все файлы .counts по генам в один общий файл `ALL.counts`.
* В качестве заголовка для каждого образца указываются не его SRR id, а c1, c2, c3 для контрольных образцов и r1, r2, r3 для перепрограммированных образцов.

```python
import pandas as pd

data = []
for name in [["SRR3414629", "r1"], ["SRR3414630", "r2"], ["SRR3414631", "r3"],
               ["SRR3414635", "c1"], ["SRR3414636", "c2"], ["SRR3414637", "c3"]]:
    data.append(pd.read_csv(f"{name[0]}.counts", sep="\t", names=[name[1]]))

data_all = pd.concat(data, axis=1)
data_all.to_csv("ALL.counts")

data_all
```
