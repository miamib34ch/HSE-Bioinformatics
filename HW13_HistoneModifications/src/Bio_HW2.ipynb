## Выравнивание на хромосому

Следует выбрать одну хромосому, потому что ресурсы google colab ограничены. В случае проблем, таких как Segmentation fault, error 139, следует выбрать хромосому меньшего размера.

Последовательность нуклеотидов хромосомы можно скачать по адресу https://hgdownload.soe.ucsc.edu/goldenPath/hg38/chromosomes/

```python
!wget https://hgdownload.soe.ucsc.edu/goldenPath/hg38/chromosomes/chr14.fa.gz
!gzip -d chr14.fa.gz
```

```python
%%time
!bowtie2-build chr14.fa chromosome_index
```

```python
!mkdir bowtie2_res
```

```python
%%time
!bowtie2  -p 2 \
          -x chromosome_index \
          -U /content/ENCFF001FEF.fastq \
          -S bowtie2_res/res_FEF.sam
```

```python
%%time
!bowtie2  -p 2 \
          -x chromosome_index \
          -U /content/ENCFF001FEG.fastq \
          -S bowtie2_res/res_FEG.sam
```

```python
%%time
!bowtie2  -p 2 \
          -x chromosome_index \
          -U /content/ENCFF001FFI.fastq \
          -S bowtie2_res/res_FFI.sam
```

## Peak calling

```python
!mkdir macs2
```

```python
!macs2 callpeak --broad -t bowtie2_res/res_FEF.sam \
    -c bowtie2_res/res_FFI.sam \
 	  -f SAM \
	  --outdir macs2 \
		-n ENCFF001FEF
```

```python
!macs2 callpeak --broad -t bowtie2_res/res_FEG.sam \
    -c bowtie2_res/res_FFI.sam \
 	  -f SAM \
	  --outdir macs2 \
    -n ENCFF001FEG
```
