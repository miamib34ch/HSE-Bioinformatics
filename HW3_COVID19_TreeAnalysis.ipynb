Выводы по деревьям:
1. Neighbor Joining: Ближе всего (заразились раньше) к SARS-CoV-1 Китай и Франция; дальше всего Новая Зеландия и Австралия, заразились позже.  
2. Maximum Likelihood: Ближе всего также Китай и Франция; а дальше всего Новая Зеландия и Австралия.
3. Maximum Parsimony: Ближе всего США; дальше всего Новая Зеландия и Австралия.
4. UPGMA: Ближе всего США и затем Нигерия; дальше всего Италия и Франция.
