```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# round number up to the second digit
def foo1(x):
    return round(x,2)

# take minimum batween number and 150, then round it
def foo2(x):
    return round(min(x,150),2)
```

```python
df1=pd.read_excel(r'/content/data.xlsx',index_col=0)
df2=pd.read_excel(r'/content/data.xlsx',index_col=0)

for name in list(df1.columns):
    df1[name] = df1[name].apply(foo1)
    
for name in list(df2.columns):
    df2[name] = df2[name].apply(foo2)
```

```python
f, ax = plt.subplots(figsize=(20,5))
ax = sns.heatmap(df1,cmap=sns.color_palette("rocket_r", as_cmap=True),annot=df1,fmt='g')
ax.set_yticklabels(df1.index.values)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(20,5))
ax = sns.heatmap(df2,cmap=sns.color_palette("rocket_r", as_cmap=True),annot=df2,fmt='g')
ax.set_yticklabels(df2.index.values)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(20,5))
colors = ['#000000','#1c001a','#290026','#40003b','#52024b','#b700cf','#ffffff']
ax = sns.heatmap(df1, cmap = colors, annot = df1, fmt = 'g')
ax.set_yticklabels(df1.index.values)
plt.show()
```
